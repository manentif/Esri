# Deploying doccano for data labeling

1. Navigate to doccano_deployment_scriptfolder.
- Run install.bat as administrator. Please be patient, the setup will take ~40 minutes to complete.
- On the command prompt, you will be asked to create your username and password for accessing Doccano.
- Once the command prompt closes, you should have Doccano running on your local system.
- Open your browser and go to https:\\localhot:8000\

# Labelling data for training with Doccano.

1.	Go to [Doccano hompage](http://localhost:8000) and login with your credentials.
2.	Create new project with project type 'Sequence labelling': 



3. Click on import data

4. Select 'JSONL' and then click on 'Select a file' and point it to the reports file (docanno_deployment_scripts\reports_labels.json). Alternatively, text documents can also be uploaded using the plain option. You can either upload one text document at a time or upload a single text file with one document per line.

5. After the file has been imported, you will see the documents loaded on the screen. 
6. Click on 'Annotate Data' from the top menu bar.

7. All the documents are pre-labelled, just 3 (document number 2,3 & 4) are intentionally left unlabeled for you to try labelling. Analyze the first labelled document and then move on to second document (use the bottom navigation bar for sifting through the docs). Mark sequences with your mouse and select the relevant title.

8.	New labels can also be created by navigating to Edit data > Labels
9.	Once all the documents have been labelled, go to 'Edit Data' > 'Export Data'
10.	Select JSON(Text-Labels)
11.	Click Download
