## Supervised and Unsupervised Machine Learning

Two of the most widely adopted machine learning methods are `supervised learning` which trains algorithms based on example input and output data that is labeled by humans, and `unsupervised learning` which provides the algorithm with no labeled data in order to allow it to find structure within its input data. Let’s explore these methods in more detail.

### Supervised Learning

In supervised learning, the example inputs that are labeled with their desired outputs are provided. The purpose of this method is for the algorithm to be able to “learn” by comparing its actual output with the “taught” outputs to find errors, and modify the model accordingly. Supervised learning therefore uses patterns to predict label values on additional unlabeled data.

Examples with supervised learning may include, (1) an algorithm may be fed data with images of sharks labeled as fish and images of oceans labeled as water. By being trained on this data, the supervised learning algorithm should be able to later identify unlabeled shark images as fish and unlabeled ocean images as water. (2) Or an algorithm using historical stock market information to anticipate upcoming fluctuations, or be employed to filter out spam emails. Note that, a common use case of supervised learning is to use historical data to predict statistically likely future events. (3) Or in supervised learning, tagged photos of dogs can be used as input data to classify untagged photos of dogs.

#### Data Augmentation and Few-Shot Learning

Data augmentation (DA), is a powerful tool and might be a big part of the solution to collect and label additional data in CV tasks. Though labeling additional samples is a time-consuming and expensive task, it does deliver better results.

Innovative approaches are needed when we need to build a classification with only one or two samples per class, and each sample is super difficult to find. Few-Shot Learning (FSL) is just one of them.

##### What is Few-Shot learning?
Few-Shot Learning (FSL) is a sub-area of machine learning. It’s about classifying new data when you have only a few training samples with supervised information. For example, workers in healthcare might have a problem with categorizing bone illnesses via x-ray photos when some rare pathologies might lack enough images to be used in the training set. This is exactly the type of problem that can be solved by building an FSL classifier.

In general, when we’re talking about FSL, it refers to `N-way-K-Shot-classification`. In which, N stands for the number of classes, and K for the number of samples from each class to train on. Also, there are two must-haves to define an FSL:

 - A training (support) set that consists of:
   - N class labels
   - K labeled images for each class (a small amount, less than ten samples per class)
 - Q query images

Now, we want to classify Q query images among the N classes. Since the N * K samples in the training set are the only examples that we have, we are facing a main problem here - there is not enough training data. To solve the FSL problems, there are generally two approaches that we can consider:

 - Data-level approach (DLA)
   - This approach is simply using additional information from a large `base dataset`. The key feature of the `base dataset` is that it doesn’t have classes that we have in our `support set` for the Few-Shot task. For example, if we want to classify a specific bird species, the `base dataset` should have images of many other birds. Also, users can produce more data themselves through data augmentation, or even `generative adversarial networks` (GANs). 
 - Parameter-level approach (PLA)
   - From the parameter-level point of view, it’s often easy to overfit on Few-Shot Learning samples, as they have extensive and high-dimensional spaces. Limiting the parameter space and use regularization and proper loss functions, can help overcome this problem, as such that the model will generalize the limited number of training samples.
   - On the other hand, model performance can be enhanced by being directed to the extensive parameter space (since we use a standard optimization algorithm, it might not give reliable results because of the small amount of training data). Hence, on the parameter-level, we train our model to find the best route in the parameter space as to give optimal prediction results. This technique is also called `Meta-Learning`.

### Unsupervised Learning

In unsupervised learning, data being input is unlabeled, so the learning algorithm is left to find commonalities among its input data. As unlabeled data are more abundant than labeled data, machine learning methods that facilitate unsupervised learning are particularly valuable.

The goal of unsupervised learning may be as straightforward as discovering hidden patterns within a dataset, but it may also have a goal of feature learning, which allows the computational machine to automatically discover the representations that are needed to classify raw data.

Unsupervised learning is commonly used for transactional data. You may have a large dataset of customers and their purchases, but as a human you will likely not be able to make sense of what similar attributes can be drawn from customer profiles and their types of purchases. With this data fed into an unsupervised learning algorithm, it may be determined that women of a certain age range who buy unscented soaps are likely to be pregnant, and therefore a marketing campaign related to pregnancy and baby products can be targeted to this audience in order to increase their number of purchases.

Without being told a “correct” answer, unsupervised learning methods can look at complex data that is more expansive and seemingly unrelated in order to organize it in potentially meaningful ways. Unsupervised learning is often used for anomaly detection including for fraudulent credit card purchases, and recommender systems that recommend what products to buy next. In unsupervised learning, untagged photos of dogs can be used as input data for the algorithm to find likenesses and classify dog photos together.

One of the main reason that KNN has been widely used in real-life scenarios, is because it is non-parametric, meaning that, it does not make any underlying assumptions about the distribution of data (as opposed to other algorithms assume e.g. a Gaussian distribution of the given data).

Another feature of KNN is that it is practicing lazy learning. Lazy learning means that the algorithm makes no generalizations, a.k.a., there is little training involved when using this method. Because of this, all of the training data is also used in testing when using KNN.

### Decision Tree Learning

Decision Trees are a non-parametric supervised learning method used for classification or regression tasks. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features. The decision rules are generally in form of if-then-else statements. The deeper the tree, the more complex the rules and fitter the model.

A decision tree is a tree-like graph with nodes representing the place where we pick an attribute and ask a question; edges represent the answers the to the question; and the leaves represent the actual output or class label. They are used in non-linear decision making with simple linear decision surface.

Decision trees classify the examples by sorting them down the tree from the root to some leaf node, with the leaf node providing the classification to the example. Each node in the tree acts as a test case (conditions as defined in if-then-else statement) for some attribute, and each edge descending from that node corresponds to one of the possible answers to the test case. This process is recursive in nature and is repeated for every subtree rooted at the new nodes.

### Deep Learning

Deep learning imitates how the human brain processes light and sound stimuli into vision and hearing. A deep learning architecture is inspired by biological neural networks and consists of multiple layers in an artificial neural network made up of hardware and GPUs.

Deep learning uses a cascade of nonlinear processing unit layers in order to extract or transform features (or representations) of the data. The output of one layer serves as the input of the successive layer. In deep learning, algorithms can be either supervised and serve to classify data, or unsupervised and perform pattern analysis.

Among the machine learning algorithms that are currently being used and developed, deep learning absorbs the most data and has been able to beat humans in some cognitive tasks. Because of these attributes, deep learning has become an approach with significant potential in the artificial intelligence space.

Computer vision and speech recognition have both realized significant advances from deep learning approaches. IBM Watson is a well-known example of a system that leverages deep learning.

#### Intro to deep learning 

Deep learning algorithms are constructed with connected layers. As shown in Fig 7, the first layer is called the Input Layer, and the last layer is called the Output Layer. All layers in between are called Hidden Layers. In order to qualify as `Deep` Learning, the network has to join neurons in more than two layers.

Each Hidden layer is composed of neurons. The neurons are connected to each other. The neuron will process and then propagate the input signal it receives the layer above it. The strength of the signal given the neuron in the next layer depends on the weight, bias and activation function.

#### Types of deep learning

`Feed Forward Neural Network` is the simplest type of artificial neural network. With this type of architecture, information flows in only one direction - forward. It means, the information’s flows starts at the input layer, goes to the `hidden` layers, and end at the output layer. The network does not have a loop, and the information stops at the output layers.

`Recurrent Neural Network` (RNN) is a multi-layered neural network that can store information in context nodes, allowing it to learn data sequences and output a number or another sequence. In short, it is an artificial neural networks whose connections between neurons include loops. RNNs are well suited for processing (time-series) sequences of inputs.

`Convolutional neural networks` (CNN) is a multi-layered neural network with a unique architecture designed to extract increasingly complex features of the data at each layer to determine the output. CNN is mostly used when there is an unstructured data set (e.g., images) and the practitioners need to extract information from it, hence well suited for perceptual tasks.

#### Neural Network

Besides brain analogies, we can also introduce neural networks appealing to Algebra. If we can recall that in linear classification we compute scores for different visual categories given the image using the formula $s=Wx$, where $W$ was a matrix and $x$ was an input column vector containing all pixel data of the image. In the case of `CIFAR-10`, $x$ is a [3072x1] column vector, and W is a [10x3072] matrix, so that the output scores is a vector of 10 class scores.

Now if we consider a neural network, it would instead compute $s=W_2 max(0,W_1 x)$. Here, $W_1$ could be, for example, a [100x3072] matrix transforming the image into a 100-dimensional intermediate vector. The function $max(0,−)$ is a non-linearity that is applied element-wise (which is a common choice among all to achieve non-linearity and simply thresholds all activations that are below zero to zero). Finally, the matrix $W_2$ would then be of size [10x100], so that we again get 10 numbers out that we interpret as the class scores. Notice that the non-linearity is critical computationally - if we left it out, the two matrices could be collapsed to a single matrix, and therefore the predicted class scores would again be a linear function of the input. The non-linearity is where we get the wiggle. The parameters $W_2$, $W_1$ are learned with stochastic gradient descent, and their gradients are derived with chain rule (and computed with backpropagation).

A three-layer neural network could analogously look like $s=W_3 max(0, W_2 max(0,W_1 x))$, where all of $W_1$, $W_2$, $W_3$ are parameters to be learned. The sizes of the intermediate hidden vectors are hyperparameters of the network and we will discuss how we can set them later. But for now, let's look into how we can interpret these computations from the neuron/network perspective.

#### Modeling a neuron

The basic computational unit of the brain is a neuron, which consists of synapses, dendrites, and axons. In the computational model of a neuron, the signals that travel along the axons (e.g. $x_0$) interact multiplicatively (e.g. $w_0 x_0$) with the dendrites of the other neuron based on the synaptic strength at that synapse (e.g. $w_0$). The idea is that the synaptic strengths (the weights w) are learnable and control the strength of influence (and its direction: excitory (positive weight) or inhibitory (negative weight)) of one neuron on another. In the basic model, the dendrites carry the signal to the cell body where they all get summed. 

Then, if the final sum is above a certain threshold, the neuron can fire, sending a spike along its axon. In the computational model, we assume that the precise timings of the spikes do not matter, and that only the frequency of the firing communicates information. Based on this rate code interpretation, we model the firing rate of the neuron with an activation function $f$, which represents the frequency of the spikes along the axon. Historically, a common choice of activation function is the sigmoid function $σ$, since it takes a real-valued input (the signal strength after the sum) and squashes it to range between 0 and 1.

#### Gradient descent

There are three key components in context of any computer vision task (e.g. image classification):

 - A (parameterized) `score function` mapping the raw image pixels to class scores (e.g. a linear function)

 - A `loss function` measures the quality of a particular set of parameters based on how well the induced scores agreed with the ground truth labels in the training data. We saw that there are many ways and versions of this (e.g. Softmax/SVM). For instance, we can recall that the linear function has the form $f(x_i,W)=W x_i$ and the SVM can be formulated as: $L = \frac{1}{N} \sum_i \sum_{j\neq y_i} \left[ \max(0, f(x_i; W)_j - f(x_i; W)_{y_i} + 1) \right] + \alpha R(W)$
 
 - `Optimization`, which is the process of finding the set of parameters $W$ that minimize the loss function.
 
Gradient, in this sense, is a way to compute the best direction along which we should change our weight vector that is mathematically guaranteed to be the direction of the steepest descent (at least in the limit as the step size goes towards zero). In one-dimensional functions, the slope is the instantaneous rate of change of the function at any point you might be interested in. The gradient is a generalization of slope for functions that don’t take a single number but a vector of numbers. Additionally, the gradient is just a vector of slopes (more commonly referred to as derivatives) for each dimension in the input space. The mathematical expression for the derivative of a 1-D function with respect its input is:

$\frac{df(x)}{dx} = \lim_{h\ \to 0} \frac{f(x + h) - f(x)}{h}$

The gradient tells us the direction in which the function has the steepest rate of increase, but it does not tell us how far along this direction we should step. However, choosing the step size (also called the `learning rate`) is one of the most important (and most headache-inducing) hyperparameter settings in training a neural network. If the step size is incremented carefully, we can expect to make consistent but very small progress (this corresponds to having a minimal or small step size). Conversely, we can choose to make a large, confident step in an attempt to descend faster, but this may not pay off (since at some point taking a bigger step gives a higher loss as we "overstep").

Now that we can compute the gradient of the loss function, the procedure of repeatedly evaluating the gradient and then performing a parameter update is called `Gradient Descent`, which is currently by far the most common and established way of optimizing Neural Network loss functions.

#### Batch, LR and other params

Training data in large-scale applications, can have on order of millions of examples. It is neither realistic or resource efficient to compute the full loss function over the entire training set in order to perform only a single parameter update. A very common approach to addressing this challenge is to compute the gradient over `batches` of the training data. For example, in current state of the art `ConvNets`, a typical batch contains 256 examples from the entire training set of 1.2 million. Such a way used to perform a parameter update is called `Mini-batch Gradient descent`, which works well with examples of which the training data are correlated.

There is one extreme case where the mini-batch contains only a single example. This process is called `Stochastic Gradient Descent (SGD)` (or also sometimes `on-line gradient descent`). This is relatively less common to see because in practice due to vectorized code optimizations it can be computationally much more efficient to evaluate the gradient for 100 examples, than the gradient for one example 100 times. Even though `SGD` technically refers to using a single example at a time to evaluate the gradient, you will hear people use the term `SGD` even when referring to mini-batch gradient descent.

The size of the mini-batch is a hyperparameter but less common to be cross-validated. It is usually based on memory constraints (if any), or set to some value, e.g. 32, 64 or 128. We use powers of 2 in practice because many vectorized operation implementations work faster when their inputs are sized in powers of 2.

## Biases, Accuracies, and Learning Rate (LR)

In the realm of machine learning, biases, accuracies, and learning rate (LR) are important concepts that are used to evaluate and improve the performance of machine learning models.

Biases refer to systematic errors in a machine learning model that cause it to consistently predict values that are different from the true values. Biases can arise from various sources, such as incomplete or biased data, inappropriate model assumptions, or limitations in the model's architecture. Biases can be reduced by increasing the complexity of the model, improving the quality and diversity of the data, or using regularization techniques that penalize complex models.

Accuracies refer to the performance of a machine learning model in terms of its ability to correctly predict values. Accuracy is typically measured as the ratio of correct predictions to the total number of predictions. However, accuracy can be misleading in some cases, such as when the classes are imbalanced or when the cost of false positives and false negatives is different. In these cases, other metrics such as precision, recall, F1-score, or AUC-ROC may be more appropriate.

LR is a hyperparameter that controls the step size of the optimization algorithm used to train a machine learning model. The LR determines how quickly or slowly the model updates its parameters during training. If the LR is too high, the model may overshoot the optimal solution and become unstable. If the LR is too low, the model may converge too slowly or get stuck in a suboptimal solution. Choosing an appropriate LR is critical for achieving good performance in machine learning models.

In summary, biases, accuracies, and learning rate are important concepts in machine learning that are used to evaluate and improve the performance of machine learning models. Understanding these concepts can help data scientists and machine learning practitioners design better models and make more informed decisions.

#### Stratification and Class Balancing

In machine learning, stratification is a technique used to ensure that the distribution of classes in the training and validation datasets is representative of the distribution in the overall dataset. Stratification is important when working with imbalanced datasets, where one or more classes are underrepresented.

`arcgis.learn` provides tools for stratifying datasets, which can be useful in several ways:

 - `Stratified Sampling`: arcgis.learn provides tools for performing stratified sampling, which ensures that the training and validation datasets contain a representative sample of all classes in the dataset. This can improve the accuracy of machine learning models, especially when working with imbalanced datasets.

 - `Stratified K-Fold Cross-Validation`: arcgis.learn provides tools for performing stratified K-Fold cross-validation, which is a technique used to evaluate the performance of machine learning models. Stratified K-Fold cross-validation ensures that each fold of the validation set contains a representative sample of all classes in the dataset.

Class balancing is another technique used in machine learning to address the issue of imbalanced datasets. Class balancing involves adjusting the distribution of classes in the training dataset to ensure that each class is represented equally. Class balancing can be done using a variety of techniques, such as oversampling, undersampling, and data augmentation.

`arcgis.learn` provides tools for performing class balancing using oversampling and undersampling techniques. `Oversampling` involves randomly duplicating instances of the minority class to balance the distribution of classes, while `undersampling` involves randomly removing instances of the majority class. arcgis.learn also provides tools for data augmentation, which can be used to generate additional training data for the minority class.

Overall, stratification and class balancing are important techniques in arcgis.learn for improving the performance of machine learning models, especially when working with imbalanced datasets.
 
#### A sample model training workflow (show_batch, LR_find, freeze and unfreeze)

`show_batch`, `LR_find`, `freeze`, and `unfreeze` are some of the commonly used methods in the arcgis.learn library, which is a high-level API for deep learning workflows in the ArcGIS platform. Here's a brief description of each method:

 - `show_batch`: This method is used to display a batch of data from a DataLoader object in arcgis.learn. It can be used to visualize the pre-processing steps applied to the data, such as resizing, normalization, and data augmentation.

 - `LR_find`: This method is used to find the optimal learning rate for a deep learning model. It works by gradually increasing the learning rate during training and monitoring the loss function. The optimal learning rate is the value at which the loss function is decreasing the most rapidly. LR_finder can help to avoid overfitting and underfitting by finding the best learning rate for the dataset and model architecture.

 - `freeze`: This method is used to freeze the weights of the earlier layers of a pre-trained model, so that only the final layers can be fine-tuned during training. Freezing the earlier layers can help to prevent the overfitting of the model and speed up the training process. Fore more, please go to [API Ref](https://developers.arcgis.com/python/api-reference/arcgis.learn.text.html#arcgis.learn.text.EntityRecognizer.freeze).

 - `unfreeze`: This method is used to unfreeze the weights of all layers of a pre-trained model, so that all layers can be fine-tuned during training. Unfreezing the earlier layers can help to improve the performance of the model, especially when working with small datasets or complex architectures. Fore more, please go to [API Ref](https://developers.arcgis.com/python/api-reference/arcgis.learn.text.html#arcgis.learn.text.EntityRecognizer.unfreeze).

Overall, these methods are useful tools for working with deep learning workflows in arcgis.learn. They can help to streamline the training process, improve the performance of the model, and optimize hyperparameters such as learning rates.

In machine learning, transfer learning is a technique where a pre-trained model is used as a starting point for a new model to solve a related problem. The pre-trained model has already learned features from a large dataset, and these features can be used as a starting point for the new model to learn from a smaller, related dataset.

When using transfer learning, the earlier layers of the pre-trained model are usually frozen, and only the later layers are fine-tuned during training. This is because the earlier layers have already learned general features that can be applied to a wide range of problems, while the later layers need to be adjusted to the specific problem and dataset.

However, in some cases, fine-tuning all layers of the pre-trained model may be necessary to achieve the best performance on the new task. This is especially true when the new dataset is small or the pre-trained model is not sufficiently similar to the new task. In such cases, unfreezing all layers and fine-tuning the entire model can help the model learn more relevant features for the new task.

Unfreezing and fine-tuning the pre-trained model can be computationally expensive and can also lead to overfitting if not done carefully. Therefore, it is important to evaluate the performance of the model on a validation set during fine-tuning and adjust the learning rate and other hyperparameters as needed.

In summary, unfreezing and fine-tuning the pre-trained model can help to improve the performance of the model on a new task, especially when the new dataset is small or the pre-trained model is not sufficiently similar to the new task.
