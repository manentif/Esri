## Computer Vision Tasks

One area of AI where deep learning has done exceedingly well is computer vision, or the ability for computers to see. The main tasks of Computer Vision include Image Classification, Object Detection, Semantic Segmentation and Instance Segmentation. 
 - `Image classification` means identifying the class to which the object belongs. For example, there are objects that belong to various classes such as dogs, cats, etc. The deep learning model will determine that the animal detected in the image belongs to the cow class with the highest probability. As shown in Fig 1, image classification is to identify that cows are in the image.
 - `Object detection` enables the detection of objects in an image and their spatial location. Bounding boxes, that are rectangles, are used to delimit the object within them. Here in Fig 1, object detection means to recognize and locate the cows.
 - As for `Semantic Segmentation`, it is able to identify similar objects in the image that belong to the same class at the pixel level. The similar objects are colored in the same way to symbolize belonging to the same class. For instance, cows are detected and located in the space without distinguishing instances (as seen in Fig 1).
 - Finally, `Instance Segmentation` recognizes the different instances given in the image with their boundaries at the deep pixel level. In this case, as what's reflected in Fig 1, all objects in the picture are identified as cows, belonging to the same class with different instances.

### Compare and Contrast CV and GIS tasks

CV and GIS tasks are both used to analyze data and create models from end to end. However, they have some key differences. CV tasks generally involve the use of machine learning algorithms and image processing techniques to analyze and classify images and videos. GIS tasks involve the use of geographic information systems to visualize, analyze, and manage geographic data.

CV tasks generally involve the use of supervised learning algorithms to classify images and videos. This includes training a model on a dataset of labeled images and videos, and then using the model to classify new images and videos. GIS tasks involve using geographic information systems to store, analyze, and visualize geographic data. This includes creating maps, visualizing data, and analyzing spatial relationships.

In summary, the difference between CV and GIS tasks include use of varied algorithms and data, and separate means of techniques and purposes.

#### Can E2E be used to help solve CV or GIS problems?

End-to-end (E2E) learning is yet another hot topic in the Deep Learning field for taking advantage of `Deep Neural Network`’s (DNNs) structure, composed of several layers, to solve complex problems. Similar to the human brain, each DNN layer (or group of layers) can specialize to perform intermediate tasks necessary for such problems. 

By definition, E2E learning refers to training a possibly complex learning system represented by a single model (specifically a DNN) that represents the complete target system, bypassing the intermediate layers usually present in traditional pipeline designs (Source: <a href="https://towardsdatascience.com/e2e-the-every-purpose-ml-method-5d4f20dafee4">TowardsDataScience.com</a>). According to Tobias Glasmachers, who evidentiated how E2E is framed in the Deep Learning context, "This elegant although straightforward and somewhat brute-force technique [E2E] has been popularized in the context of deep learning. It is a seemingly natural consequence of deep neural architectures blurring the classic boundaries between learning machine and other processing components by casting a possibly complex processing pipeline into the coherent and flexible modeling language of neural networks."

The benefits of practicing E2E learning involves:
 - replacing the pipelined chain (in which each module has to be optimized separately under different criteria) for a single Neural Network, allowing the use of a single optimization criterion for enhancing the system.
 - it is possible to design a model that performs well without deep knowledge about the problem, despite its complexity. According to <a href="Natural language processing (almost) from scratch.” Journal of machine learning research 12.Aug (2011): 2493–2537">Ronan Collobert et al.</a>,  a unified Neural Network architecture and an appropriate learning algorithm can be used to avoid task-specific engineering and lots of prior knowledge.

### Difference between object/pixel classification, and object detection

Object/Pixel Classification is the process of assigning a label to each pixel in an image. This label can be a class such as “building”, “road”, or “tree”. The goal of this process is to identify the objects in the image. Whereas, Object Detection is the process of locating and identifying objects in an image, which involves identifying the location of an object in the image, as well as its size and shape. The goal of this process is to identify and locate multiple objects in an image.

#### The goal of object classification

Object Classification is a type of computer vision task in which objects are identified and classified into pre-defined categories. The goal of object classification is to correctly identify and classify objects within an image or video frame in order to better understand the content of the image or video. By accurately classifying objects, it can be easier to detect patterns and make predictions about the content of the image or video.

#### Required inputs (input raster, and the feature class)

The required inputs to perform an object classification typically include a feature class or raster, training data, and classification rules. The feature class or raster is the dataset containing the features to be classified, such as land cover, roads, buildings, etc. The training data is the labeled data used to train the classification algorithm. The classification rules are the parameters used to define how the features should be classified, such as thresholds, distance metrics, etc.

The goal of object classification is to determine the class of each feature, such as a building. For example, you can use it to determine if a building is damaged after a natural disaster. Object classification requires the following inputs:
 - An input raster that contains the spectral bands
 - A feature class that defines the location (for example, an outline or a bounding box) of each feature

There are two major steps in feature classification. We first export training samples based on the geographical extent of each feature. Once the training samples are exported, it can be used as the training input for the deep learning based classification algorithm to train a feature classifier.

You can solve object classification through Convolutional Neural Networks (CNN). There are many CNN-based image classification algorithms. Most algorithms have a backbone that uses CNN architecture, such as Resnet, LeNet-5, AlexNet, or VGG 16, which is then followed by a softmax layer.

Object classification uses the Feature Classifier model type to train a model. A model definition file containing trained model information hence comes into the picture. The model can be trained using the `Train Deep Learning Model` tool or by a third-party training software such as `TensorFlow`, `PyTorch`, or `Keras`. The model definition file can be an `Esri model definition JSON file (.emd)` or a `deep learning model package (dlmp)`, and it must contain the path to the Python raster function to be called to process each object and the path to the trained binary deep learning model file.

#### What is CNN, backbone, and classification head?

A `convolutional neural network (CNN)` is a type of deep learning neural network that is typically used for image recognition, natural language processing, and other computer vision tasks. CNNs are composed of an input and an output layer, as well as multiple hidden layers. The hidden layers are composed of convolutional layers, pooling layers, and fully connected layers. CNNs are designed to recognize patterns in images and use those patterns to classify images or detect objects.

A `backbone` in deep learning is a type of neural network architecture that is used as the basis of a more complex network. It is typically composed of convolutional layers and pooling layers, and is often used to extract features from input data. Backbones are also used to reduce the complexity of a model, allowing for faster training and better performance. We can think of it as a kind of preconfigured formula or template.

A `classification head` is a term used in artificial intelligence (AI) to describe a type of model used to classify data. It is a part of a larger AI model that is responsible for making predictions based on the input data. The classification head uses a variety of techniques, such as neural networks and decision trees, to analyze the input data and make a prediction.

##### More with CNN

CNNs are a type of deep learning algorithm that are used to process data with a grid-like topology. CNNs are a type of deep learning algorithm that is used to process data that has a spatial or temporal relationship. CNNs are similar to other neural networks, but they have an added layer of complexity due to the fact that they use a series of convolutional layers. Convolutional layers are an essential component of Convolutional Neural Networks (CNNs). Fig 4 below represents a typical CNN architecture.

Layers include,
 - `Convolutional layer`: Convolutional layers are made up of a set of filters (also called kernels) that are applied to an input image. The output of the convolutional layer is a feature map, which is a representation of the input image with the filters applied. Convolutional layers can be stacked to create more complex models, which can learn more intricate features from images. 
 - `Pooling layer`: Pooling layers are a type of convolutional layer used in deep learning. Pooling layers reduce the spatial size of the input, making it easier to process and requiring less memory. Pooling also helps to reduce the number of parameters and makes training faster. There are two main types of pooling: max pooling and average pooling. Max pooling takes the maximum value from each feature map, while average pooling takes the average value. Pooling layers are typically used after convolutional layers in order to reduce the size of the input before it is fed into a fully connected layer. 
 - `Fully connected layer`: Fully-connected layers are one of the most basic types of layers in a convolutional neural network (CNN). As the name suggests, each neuron in a fully-connected layer is Fully connected- to every other neuron in the previous layer. Fully connected layers are typically used towards the end of a CNN- when the goal is to take the features learned by the previous layers and use them to make predictions. For example, if we were using a CNN to classify images of animals, the final Fully connected layer might take the features learned by the previous layers and use them to classify an image as containing a dog, cat, bird, etc. 
 
CNNs are often used for image recognition and classification tasks. For example, CNNs can be used to identify objects in an image or to classify an image as being a cat or a dog. CNNs can also be used for more complex tasks, such as generating descriptions of an image or 1 identifying the points of interest in an image. CNNs can also be used for time-series data, such as audio data or text data. CNNs are a powerful tool for deep learning, and they have been used to achieve state-of-the-art results in many different applications.

####  CNN-based image classification algorithms

In current days, the most popular CNN-based image classification algorithms include:

 - 1. `AlexNet`: AlexNet is a convolutional neural network (CNN) that revolutionized image classification by becoming the first to demonstrate the potential of deep learning in the field of computer vision. AlexNet was trained to recognize over 1000 object categories by having millions of images as inputs. It was the winning entry in the `2012 ImageNet competition`, which is an annual competition for computer vision tasks. AlexNet is made up of eight layers and consists of 11×11, 5×5, 3×3, and 1×1 convolutional filters, a max-pooling layer, and three fully connected layers. It also uses the ReLU activation function to reduce complexity and improve performance. AlexNet is commonly used for image classification tasks such as object detection, face recognition, scene classification, and more.

 - 2. `VGGNet`: VGGNet is a convolutional neural network (CNN) architecture developed by the `Visual Geometry Group (VGG)` at the University of Oxford. It is used for image classification, object recognition, and object detection tasks. The VGGNet architecture is made up of multiple convolutional layers, which are followed by a few fully connected layers. It is used to identify objects in images by applying a series of filters to the input image to extract key features from the image. The extracted features are then used to classify the image into the desired object category. 

 - 3. `GoogLeNet`: GoogleNet (also known as `Inception`) is an image classification convolutional neural network (CNN) developed by Google for the `ImageNet Large Scale Visual Recognition Challenge (ILSVRC)` in 2014. It uses a deep neural network architecture that consists of 22 layers and won the challenge with a top 5 error rate of 6.67%. The network architecture was designed to leverage the strengths of existing architectures and to introduce new features, such as the inception module, which is a combination of multiple convolutional and pooling layers fed into a single neuron. GoogleNet is used in a variety of image classification tasks, such as object detection, image segmentation, and image recognition. It has been used in many applications such as image captioning, facial recognition, and autonomous vehicles. 

 - 4. `ResNet`: ResNet (`Residual Network`) is a deep learning architecture that was developed by Microsoft Research for image classification and recognition. It is a convolutional neural network (CNN) that is used for image recognition and classification by taking an image as input and outputting a class label or class probability for the image. ResNet is designed to be used in any image analysis task, including image classification, object detection, and object segmentation. The main feature of ResNet is the use of `residual learning`, which allows the network to learn the identity mapping (predicting an input image as itself) and then use that knowledge to better recognize other objects in the image. This helps the network learn more quickly and accurately. 

 - 5. `MobileNet`: MobileNet is a type of convolutional neural network (CNN) that was developed by Google in 2017. It is specifically designed for efficient mobile vision applications, which means it is optimized for running on resource-constrained devices such as smartphones and other mobile devices. The key innovation of MobileNet is its use of depthwise separable convolutions. This involves splitting the standard convolution operation into two separate layers: a depthwise convolution and a pointwise convolution. The depthwise convolution applies a single filter to each input channel, while the pointwise convolution applies a 1x1 convolution to combine the outputs of the depthwise convolution. This approach reduces the computational cost of the convolution operation while maintaining a similar level of accuracy. 

 - 6. `SqueezeNet`: SqueezeNet is a deep learning architecture designed for efficient image classification. It was developed by researchers at DeepScale, University of California, Berkeley, and Stanford University. SqueezeNet is a deep convolutional neural network that is 50 times smaller in size and requires significantly fewer computations than AlexNet, one of the largest CNNs. It is designed to be computationally efficient, so it can be used for image classification tasks on mobile or embedded devices where there is limited memory or processing power. SqueezeNet uses a combination of convolution, pooling, and fire modules to classify images. The fire module consists of a "squeeze" convolution layer to reduce the number of input channels, followed by an "expand" layer to increase the number of output channels. The output of the fire module is then passed to a max pooling layer that reduces the resolution of the output. Finally, the output of the pooling layer is passed to a fully connected layer to make the final predictions. 

 - 7. `DenseNet`: DenseNet is a convolutional neural network architecture used for image classification. It is based on the principle of connecting each layer to every other layer in a network. This connection enables features to be more readily propagated between layers, allowing the network to learn more complex features and relationships. DenseNet is considered to be a more efficient approach to image classification as it reduces the number of parameters and computations required to train a model. It also makes the model more robust to varying input sizes and improves the accuracy of the model on small datasets.

 - 8. `LeNet`: LeNet is a convolutional neural network (CNN) that was developed to recognize handwritten digits. It was developed in the 1990s by Yann LeCun and has become the most widely used architecture for image classification. LeNet is composed of 7 layers, including convolutional, pooling, and fully connected layers. It uses various techniques, such as parameter sharing and pooling, to reduce the amount of parameters needed for training, thus making it more efficient and accurate. LeNet is used for image classification tasks such as recognizing handwritten digits, medical imaging, face recognition, and object recognition.

There are many algorithms or architectures emerging these years, and deciding which one is suitable for a specific task is not easy. That's when metrics, or evaluation benchmarks, come into the picture. For instance, `TIMM`, which stands for `Tensorflow Image Metrics Module`, is an open-source library designed to help with the development and evaluation of image classification models. It provides a suite of metrics, such as accuracy, F1 score, precision, recall, and more, that can be used to measure the performance of an image classification model. It also provides easy-to-use APIs that allow developers to quickly and easily evaluate their models and track their progress as they iterate on their models. In addition to its use for model evaluation, TIMM can also be used to visualize the results of the classification models and to find areas for improvement.

##### Models' inference speed, accuracy, model size, and their relationships

The inference speed of a CNN model depends on several factors, including the type of model, the size of the dataset, the hardware used to run the model, and the complexity of the model. Generally speaking, CNN models can range from being very fast to very slow depending on the factors mentioned above.

A CNN model's accuracy can be defined as the percentage of correctly classified images by the model. It is usually calculated by dividing the number of correctly classified images by the total number of images in the test set.

The model size of a CNN model depends on the number of layers and the size of each layer. A CNN model typically consists of convolutional layers, pooling layers, fully connected layers, and output layers. The number of parameters in the model can range from a few thousand to several million.

### Model explainability using gradcam

Model explainability is the ability of a machine learning model to explain its prediction and reasoning in a way that is understandable to humans. This is important as it helps to understand how the model works, why it makes certain decisions, and how it can be improved. It also helps to establish trust in a system and gain an understanding of any ethical implications.

GradCAM (Gradient-weighted Class Activation Mapping) is a method used to visualize which parts of an image were most important to a model when it made its classification. This allows us to understand more about how the model works and how it comes to its decisions. By visually highlighting the most important areas of an image, GradCAM makes it easier to explain the rationale behind the model’s predictions. Additionally, it can be used to identify areas where the model may be making mistakes and can suggest improvement strategies.

For example, if you have a model that classifies dogs, a GradCAM attention map could be used to highlight the parts of the image that the model is paying most attention to when it comes to making its predictions. This could be the eyes, ears, or other parts of the dog that the model is paying attention to.

#### Why is model explainability important?

Model explainability is important in 2D computer vision tasks for several reasons:

 - Debugging: When developing a computer vision model, it is important to understand how it is making decisions in order to identify and debug any errors or biases. Model explainability allows developers to see how the model is processing and interpreting the input data.

 - Trust and accountability: In many applications of computer vision, such as medical imaging or autonomous vehicles, the decisions made by the model can have significant consequences. In these cases, it is important to have a clear understanding of how the model arrived at its decisions in order to build trust and ensure accountability.

 - Compliance with regulations: In some industries, such as finance and healthcare, there are regulations that require models to be explainable in order to ensure fairness, transparency, and accountability.

 - Insights and improvements: By understanding how a model is making decisions, developers can gain insights into the strengths and weaknesses of the model and identify areas for improvement.

Overall, model explainability is crucial for ensuring the reliability and transparency of computer vision models, particularly in applications where the consequences of errors or biases could be significant.
