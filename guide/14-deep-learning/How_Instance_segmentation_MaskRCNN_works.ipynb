## Semantic and Instance Segmentation

There is two different method for segmentation (instance and semantic segmentation) used in computer vision or in deep learning. Let's understand the difference between these two methods.

Semantic segmentation is the technique in which only one-pixel value is associated to each different class while in the instance segmentation unique pixels value is assigned for each instance of the same class.

# Instance Segmentation using MaskRCNN

MaskRCNN is state of the art model for instance segmentation, developed on top of Faster RCNN. Faster RCNN is the proposal-based object detection model (return bounding boxes of each object and its class label with confidence of being that class) which is widely used for best accuracies.

# Model Architecture

To understand MaskRCNN let's first discus architecture of Faster RCNN. FasterRCNN works in two stage.

# Stage 1

The first stage consists of two networks, backbone (ResNet, VGG, Inception, etc..) and region proposal network. These networks run once per image to give set of region proposals. Region proposals is the region in feature map which contains object.

# Stage 2

In the second stage, we will predict bounding boxes and object class for each of the proposed region obtained in stage 1. Each proposed region can be of different size and fully connected layers in the networks always needed fixed size vector to make predictions, so we will make the proposed region to fixed size by use of either RoI pool(which is very similar to MaxPooling) or RoI align method.

As we have discussed MaskRCNN is developed over Faster RCNN with minor modification. In the second stage of Faster RCNN, predictions of binary mask for each RoI is take place along with object class and bounding boxes by adding one more branch which is fully convolutional network.

# Mask Representation

A mask is the spatial representation of object's present in an image. In MaskRCNN, each object will have its own binary mask, the pixels corresponding to ground is set to be zero while the pixels corresponding to the object is set to be one. So, mask will contain the actual shape of the object.
