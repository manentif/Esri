# License Management

### Setting up

```python
from arcgis.gis import GIS
online_gis = GIS(profile='ds24_online')
online_gis
```

```python
current_user = online_gis.users.me
current_user
```

```python
user_manager = online_gis.users
```

### Display all license types assigned to the org

```python
all_licenses = online_gis.admin.license.all()
all_licenses
```

### Assign a license

```python
city_engine_license = online_gis.admin.license.get("arcgis cityengine")
city_engine_license
```

```python
# list current assignments of this license
city_engine_license.all()
```

```python
# show available entitlements for this license
city_engine_license.properties.provision.orgEntitlements.entitlements
```

```python
city_engine_license.assign('jayber_ds24', 'cityEngine')
```

```python
# verify the assignment
city_engine_license.all()
```

```python
city_engine_license.report
```

```python
# revoke all entitlements from this user
city_engine_license.revoke('jayber_ds24', '*')
```

```python
# verify revocation
city_engine_license.all()
```

```python
pro_license = online_gis.admin.license.get("arcgis pro")
pro_license
```

```python
pro_license.all()
```

```python
pro_license.report
```

```python
all_users = user_manager.search('*')
# get a random user, right in the middle of the list
test_user = all_users[int(len(all_users)/3)]
test_user
```

```python
# assign ArcGIS Pro Advanced
# take a look at the report above for the available Pro entitlements
pro_license.assign(test_user.username, 'desktopAdvN')
```
